```python
import requests 
import json 

url = f'https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime={start_date}&endtime={end_date}'


response = requests.get(url)

if response.status_code == 200:
    print("Connexion avec succès")
    #Recuperer les features : Each element(json object) in «features » is a single Earthquake event.
    data = response.json()["features"]

    file_path =f"/lakehouse/default/Files/{start_date}_earthquake_data.json"

    with open(file_path, "w") as file:
        json.dump(data,file,indent=4)
        print("Data successfully load")

else : 
    print("Erreur de connexion : ", response.status_code)
```
